```python
import matplotlib.pyplot as plt

import astro30m

dir(astro30m)

env_name = "pdr1"
```

```python
av = astro30m.load_phys_quantity("dust-av", 2020).log(10)
av0 = astro30m.view_region(av, "old")

av1 = astro30m.view_region(av, "horsehead")
av2 = astro30m.view_region(av, "pdr1")

plt.figure(dpi=200)

plt.subplot(1, 3, 1)
av0.plot()

plt.subplot(1, 3, 2)
av1.plot()

plt.subplot(1, 3, 3)
av2.plot()

plt.show()
```
