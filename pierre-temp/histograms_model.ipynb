```python
import os
import sys

import numpy as np
import matplotlib.pyplot as plt

sys.path.insert(0, os.path.join("..", "paper-1"))
from pdr_util import simulate, get_physical_env
```

```python
n_samples = 100_000
bounds = get_physical_env("horsehead")
obstime = 4 * 0.1

df = simulate(
    n_samples,
    bounds,
    obstime,
    noise=True
)
```

```python
linename = "13c_o_j1__j0" 

intens = df[linename]
intens = np.log10(intens)

plt.hist(intens, bins=50, density=True)

plt.yscale("log")
plt.title(linename)
plt.xlim([-14, -7])

plt.show()
```

```python
linename = "hcop_j1__j0" 

intens = df[linename]
intens = np.log10(intens)

plt.hist(intens, bins=50, density=True, cumulative=True)

# plt.yscale("log")
plt.title(linename)
plt.xlim([-15, -7])

plt.axvline(-10, color="tab:red")
plt.axvline(-9, color="tab:red")

plt.grid()

plt.show()
```

```python
linename = "hcop_j1__j0" 

intens = df[linename]
intens = np.log10(intens)

plt.hist(intens, bins=50, density=True)

plt.title(linename)
plt.xlim([-15, -7])

plt.axvline(-10, color="tab:red")
plt.axvline(-9, color="tab:red")

plt.grid()

plt.show()
```

```python
intens = df["Avmax"]
intens = np.log10(intens)

plt.hist(intens, bins=50, range=(np.log10(3), np.log10(6)),density=True)

plt.yscale("log")
plt.title('av')
# plt.xlim([-14, -7])

plt.show()
```

```python
intens = df["radm"]
intens = np.log10(intens)

plt.hist(intens, bins=50, range=(np.log10(1e1), np.log10(10**2.4)),density=True)

plt.yscale("log")
plt.title('radm')
# plt.xlim([-14, -7])

plt.show()
```

```python
intens = df["hcop_j1__j0"]
intens = np.log10(intens)

noise = -9.598276046590223 # HCO+

snr = intens-noise

#

plt.hist(snr, bins=50,density=True)
plt.xlim([-4, 2])

plt.show()
```

```python
plt.scatter(
    np.log10(df["Avmax"]), np.log10(df["radm"]),
    c=np.log10(df["hcop_j1__j0"])-noise,
    s=2
)
plt.xlabel("Av")
plt.ylabel("G0")
plt.colorbar()
plt.show()
```
